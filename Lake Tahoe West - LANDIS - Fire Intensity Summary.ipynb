## Processing LANDIS Fire Intensity raster data

### Convert ASCII to ESRI Grid and combine into one file geodatabase per scenario

```python
import arcpy, os

# Set the environment variables
arcpy.env.overwriteOutput = True
sr = arcpy.SpatialReference(26911)

# Set local variables
scenario1 = r"C:/GIS/PROJECT/LakeTahoeWest/Data/Download/USFS/LANDIS/Scenario_1_FireIntensity"
scenario2 = r"C:/GIS/PROJECT/LakeTahoeWest/Data/Download/USFS/LANDIS/Scenario_2_FireIntensity"
scenario3 = r"C:/GIS/PROJECT/LakeTahoeWest/Data/Download/USFS/LANDIS/Scenario_3_FireIntensity"
scenario4 = r"C:/GIS/PROJECT/LakeTahoeWest/Data/Download/USFS/LANDIS/Scenario_4_FireIntensity"
# replicate folders to iterate over
rep1 = "fireintensity_r1"
rep2 = "fireintensity_r2"
rep3 = "fireintensity_r3"
rep4 = "fireintensity_r4"
rep5 = "fireintensity_r5"
rep6 = "fireintensity_r6"
rep7 = "fireintensity_r7"
rep8 = "fireintensity_r8"
rep9 = "fireintensity_r9"
rep10 = "fireintensity_r10"

# lists for workspace path building
scenarios = [scenario1, scenario2, scenario3, scenario4]
reps = [rep1, rep2, rep3, rep4, rep5, rep6, rep7, rep8, rep9, rep10]

# output path to file geodatabase
outgdb = r"C:\GIS\PROJECT\LakeTahoeWest\Data\Scenario2_FireIntensity.gdb"

## Reset the workspace manually for now ##
for rep in reps:
    arcpy.env.workspace = os.path.join(scenario2,rep)
    # Get a list of rasters from the workspace
    rasters = arcpy.ListRasters("*", "ALL")
    # loop through raster list 
    for raster in rasters:
        # create describe object of each raster
        desc = arcpy.Describe(raster)
        inras = desc.CatalogPath
        year = str(desc.baseName[0:-4].split('-')[-1])
        # name = str(desc.baseName[0:-4].split('-')[-2])
        rep  = str(rep.split('_')[-1])
        # limit the loop to fire intensity rasters
        if (int(year) <=100):
            outras = (outgdb + "\\fireintensity_" + rep + "_" + year)
            arcpy.ASCIIToRaster_conversion(inras, outras)
            arcpy.DefineProjection_management(outras, sr)
            dsc = arcpy.Describe(outras)
            print ("Created: "+ dsc.baseName)
        # skip other rasters in the folder
        else:
            pass
```

```python
import arcpy

rep1 = "/replicate1/scrapple-fire"
rep2 = "/replicate2/scrapple-fire"
rep3 = "/replicate3/scrapple-fire"
rep4 = "/replicate4/scrapple-fire"
rep5 = "/replicate5/scrapple-fire"
rep6 = "/replicate6/scrapple-fire"
rep7 = "/replicate7/scrapple-fire"
rep8 = "/replicate8/scrapple-fire"
rep9 = "/replicate9/scrapple-fire"
rep10 = "/replicate10/scrapple-fire"

#replicates = (rep2, rep3, rep4, rep5, rep6, rep7, rep8, rep9, rep10)
#for rep in replicates:
arcpy.env.workspace = (r"C:/GIS/PROJECT/LakeTahoeWest/Data/Download/USFS/LANDIS/Scenario3_ArcGIS/replicate2/scrapple-fire")
rasters = arcpy.ListRasters("*", "ALL")

for raster in rasters:
    # create describe object of each raster
    desc = arcpy.Describe(raster)
    
    # limit the loop to fire intensity rasters
    if raster is (desc.baseName[0:-4].split('-')[-2] == "intensity"):
        ##Build attribute table for single band raster dataset
        ##Overwrite the existing attribute table file
        arcpy.BuildRasterAttributeTable_management(raster, "Overwrite")
        
        # Set local variables
        fieldName = "RasterName"
        # set the field to the file name -.img
        expression = "str(desc.baseName[0:-4])"
        # Add field for Raster Name
        arcpy.AddField_management(raster, fieldName, "TEXT", field_length=55)
        # Execute CalculateField 
        arcpy.CalculateField_management(raster, fieldName, expression, "PYTHON3")
    
        # identify modle year and fill in attribute
        fieldName = "Year"
        # set the field to the file name -.img
        mystring = str(desc.baseName[0:-4])
        expression = "mystring.split('-')[-1]"
        # Add field for Raster Name
        arcpy.AddField_management(raster, fieldName, "SHORT")
        # Execute CalculateField 
        arcpy.CalculateField_management(raster, fieldName, expression, "PYTHON3")
        
        print ("Attribute table built for: ".format(desc.baseName[0:-4]))
```
